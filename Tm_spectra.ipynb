```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cufflinks as cf
```

```python
Tm = pd.read_csv('Tm.asc', delimiter=r'\s+', header=None)
Ne = pd.read_csv('Ne.asc', delimiter=r'\s+', header=None)
measured = pd.read_csv('Tm_measured.txt', header=None,delimiter='\t')
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(measured[0], (measured[1]-min(measured[1]))/max(measured[1]), linestyle='-', color='black', markersize=2, label = 'measured')
plt.plot(Tm [0], Tm [1], linestyle='--', color='b', markersize=2, label = 'Tm (NIST)')
plt.plot(Ne [0], Ne [1], linestyle='--', color='r', markersize=2, label = 'Ne (NIST)')

# Adding labels and title
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (a.u.)')
plt.xlim(550, 800)
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go

# Create traces
trace1 = go.Scatter(x=measured[0], y=(measured[1]-min(measured[1]))/max(measured[1]), mode='lines+markers', name='measured')
trace2 = go.Scatter(x=Tm[0], y=Tm[1], mode='lines+markers', name='Tm (NIST)')
trace3 = go.Scatter(x=Ne[0], y=Ne[1], mode='lines+markers', name='Ne (NIST)')

# Create the figure
fig = go.Figure()

# Add the traces to the figure
fig.add_trace(trace1)
fig.add_trace(trace2)
fig.add_trace(trace3)

# Add title and labels
fig.update_layout(title='Ne-Tm lamp spectrum compared to NIST datasets',
                  xaxis_title='Relative intensity (a.u.)',
                  yaxis_title='Wavelength (nm)')

# Show the figure
fig.show()
```
