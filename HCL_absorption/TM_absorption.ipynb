```python
file = './TmI/12.csv'
df = pd.read_csv(file)#, delimiter=r'\s+', header=None)
df = df.drop(index=0)  # The index of the first row is 0

# Plot CH1 and CH4 columns
plt.figure(figsize=(10, 6))
plt.plot(df['CH1'], label='CH1')
plt.plot(df['CH4'], label='CH4')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Plot of CH1 and CH4')
plt.legend()

# Show the plot
plt.show()
```
